# Iterativno stvaranje mreže prijateljstva

```python
import matplotlib.pyplot as plt
# Create the graph and set edge weights
G = nx.Graph()
for index, row in df.iterrows():
    source = row['Ime i Prezime']
    target1 = row['Unutar vlastite studijske godine, napiši Ime i Prezime osobe s kojom si u najboljim odnosima.']
    target2 = row['Napiši još jedno Ime i Prezime osobe s kojom si u odličnim odnosima.']
    G.add_edge(source, target1, weight=5)
    G.add_edge(source, target2, weight=1)

print(G.nodes)
print(G.edges)

```

# Iscrtavanje grapha korištenjem "Spring layouta" i Networkx

```python
# Define the layout algorithm
pos = nx.spring_layout(G, k=0.8)

# Draw the nodes, edges, and edge weights
nx.draw_networkx_nodes(G, pos, node_size=200)
nx.draw_networkx_edges(G, pos, width=1)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'), font_size=5)
nx.draw_networkx_labels(G, pos, font_size=11)

# Set the plot size and display
plt.figure(figsize=(55,15))
plt.axis('off')
plt.show()
```

# Iscrtavanje grafa korištenjem interaktivne knjižnice Plotly

```python
import plotly.graph_objects as go
```

```python
def plotly_graph(df, G, k, naziv):
    # Set node colors based on columns in the DataFrame
    node_colors = []
    for node in G.nodes():
        if node in df.loc[:, 'Napiši Ime i Prezime profesora koji ti je najinspirativniji.'].unique().tolist():
            node_colors.append('blue')
        elif node in df.loc[:, 'Napiši predmet koji ti je najinspirativniji.'].unique().tolist():
            node_colors.append('green')
        elif node in df.loc[:, 'Napiši hobi s kojem bi se htio baviti'].unique().tolist():
          node_colors.append('violet')
        else:
            node_colors.append('red')

    # Calculate the degree of each node
    degree = G.degree()
    node_degrees = [x[1] for x in degree]

    # Compute the positions of the nodes in the graph using the spring layout algorithm
    pos = nx.spring_layout(G, k=k)

    # Create an empty Plotly Figure
    fig = go.Figure()

    # Add the edges to the graph
    for source, target, weight in G.edges(data='weight'):
        fig.add_trace(go.Scatter(x=[pos[source][0], pos[target][0]], y=[pos[source][1], pos[target][1]],
                                 mode='lines', line=dict(width=weight, color='black')))

    # Add the nodes to the graph
    node_trace = go.Scatter(x=[], y=[], text=[], mode='markers+text', textposition='top center',
                            hoverinfo='text', marker=dict(color='red', size=10))
    for node in G.nodes():
        x, y = pos[node]
        node_trace['x'] += tuple([x])
        node_trace['y'] += tuple([y])
        node_trace['text'] += tuple([node])

    # Set the node colors and sizes based on the previously computed values
    node_trace['marker']['color'] = node_colors
    node_trace['textfont_size'] = [x[1]*5 for x in degree]
    node_trace['marker']['size'] = [x[1]*10 for x in degree]

    node_trace['customdata'] = node_degrees
    # Add hovertemplate to display name and degree on hover
    node_trace['customdata'] = node_degrees
    node_trace['hovertemplate'] = '<b>%{text}</b><br>Degree: %{customdata}<extra></extra>'



    # Add the node trace to the figure
    fig.add_trace(node_trace)

    # Update the layout of the graph
    fig.update_layout(title=naziv, title_x=0.5, showlegend=False, hovermode='closest',
                     margin=dict(b=20, l=5, r=5, t=40), xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                     yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))

    # Show the figure
    fig.show()
```

```python
plotly_graph(df, G, 0.3, 'Mreža prijateljstva')
```

# Izračunavanje centralnosti čvorova u mreži
## Stupanj centralnosti
Stupanj centralnosti u grafu je mjera koja se koristi za procjenu koliko je jedan čvor (cvor) u grafu važan u odnosu na druge čvorove. Stupanj centralnosti mjeri koliko čvorova izravno povezano s nekim čvorom, odnosno koliko susjeda ima taj čvor.

U neusmjerenom grafu, stupanj centralnosti čvora je jednostavno broj susjeda tog čvora. U usmjerenom grafu postoje dva tipa stupnja centralnosti - ulazni stupanj centralnosti i izlazni stupanj centralnosti. Ulazni stupanj centralnosti čvora je broj čvorova koji pokazuju prema tom čvoru, dok je izlazni stupanj centralnosti čvora broj čvorova prema kojima taj čvor pokazuje.

Stupanj centralnosti se često koristi u socijalnim mrežama za identificiranje pojedinaca koji su najpovezaniji s drugim članovima mreže. Također se može primijeniti u raznim drugim područjima, poput prometa u cestovnom mrežama, računalnoj znanosti, i drugdje.
