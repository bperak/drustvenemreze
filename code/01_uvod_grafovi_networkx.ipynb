U ovoj bilježnici gradimo jednostavan graf (čvorovi i veze) i vizualiziramo ga pomoću NetworkX. Koristimo isti primjer kao u sadržaju: trokut (Ana–Bruno–Carla) i „most” prema Davidu i Evi.

## Korak 1: Izgradnja grafa

**Što radimo:** Gradimo neusmjereni graf s čvorovima (Ana, Bruno, Carla, David, Eva) i veza koje odgovaraju primjeru iz sadržaja: trokut + most. Fiksni layout (`spring_layout`, seed=42) koristit ćemo u svim crtežima.

*(Definicije: [pogl. 1 — Čvorovi i veze](../content/01_uvod_drustvene_mreze_teorija_grafova.md#definicije). Primjeri u sadržaju: [Primjer u stvarnom svijetu](../content/01_uvod_drustvene_mreze_teorija_grafova.md#primjer-u-stvarnom-svijetu).)*

```python
import networkx as nx
import matplotlib.pyplot as plt

# Korak 1: Izgradnja grafa — isti primjer kao u sadržaju (trokut + most)
# Čvorovi s čitljivim imenima; veze: Ana–Bruno–Carla trokut, Bruno–David–Eva most
G = nx.Graph()
G.add_edges_from([
    ("Ana", "Bruno"), ("Bruno", "Carla"), ("Ana", "Carla"),  # trokut
    ("Bruno", "David"), ("David", "Eva")                      # most
])
pos = nx.spring_layout(G, seed=42)  # Zajednički layout za sve crteže

print("Broj čvorova:", G.number_of_nodes())
print("Broj veza:", G.number_of_edges())
print("Stupnjevi čvorova:", dict(G.degree()))
```

## Korak 2: Vizualizacija grafa

**Što radimo:** Crtamo graf s jednakim veličinama čvorova; naslov objašnjava strukturu (trokut + most). Isti `pos` osigurava konzistentan izgled.

*(Vizualizacija osnova: [pogl. 1](../content/01_uvod_drustvene_mreze_teorija_grafova.md).)*

```python
# Korak 2: Vizualizacija — osnovna struktura (trokut + most)
plt.figure(figsize=(7, 4))
nx.draw(G, pos, with_labels=True, node_color='lightblue', font_weight='bold', node_size=600)
plt.title("Graf društvene mreže: trokut (Ana–Bruno–Carla) + most (Bruno–David–Eva)")
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Korak 3: Osnovne statistike (stupnjevi čvorova)

**Što radimo:** Prikazujemo broj veza po čvoru (stupanj) u tablici — prva mjera koju ćemo u pogl. 3 proširiti drugim mjerama centralnosti.

*(Definicije: [pogl. 1 — graf, čvorovi, veze](../content/01_uvod_drustvene_mreze_teorija_grafova.md#definicije).)*

```python
import pandas as pd
stupnjevi = dict(G.degree())
tablica = pd.DataFrame([
    {"Čvor": cvor, "Stupanj": d} for cvor, d in stupnjevi.items()
])
print(tablica.to_string(index=False))
```

## Zaključak

Graf ima 5 čvorova i 6 veza; vizualizacija (Korak 2) i tablica stupnjeva (Korak 3) ilustriraju strukturu iz sadržaja (trokut + most). U poglavljima 2 i 3: prikupljanje podataka i računanje ostalih mjera centralnosti na istom tipu grafa.
