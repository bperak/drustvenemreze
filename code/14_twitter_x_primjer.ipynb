Bez stvarnog API poziva: gradimo mrežu retweetova iz liste uređenih parova (korisnik, retweetirao). Za stvarne podatke koristiti X API v2 i .env za ključeve.

```python
import networkx as nx
# Mock retweetovi: (tko_retweetira, koga)
edges = [('user1','news'), ('user2','news'), ('user3','politician'), ('user1','politician'), ('user4','news')]
D = nx.DiGraph()
D.add_edges_from(edges)
print('Broj čvorova (korisnici/računi):', D.number_of_nodes())
print('Broj retweetova:', D.number_of_edges())
print('In-degree (tko je najviše retweetiran):', dict(D.in_degree()))
```
