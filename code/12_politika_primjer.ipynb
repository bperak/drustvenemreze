Mock mreža "retweetova": čvorovi = akteri, usmjereni bridovi = A retweetira B.

```python
import networkx as nx
D = nx.DiGraph()
D.add_edges_from([('M1','P1'),('M2','P1'),('M1','P2'),('G1','M1'),('G2','M2')])
in_d = dict(D.in_degree())
print('In-degree (tko je najviše retweetiran):', in_d)
print('Čvor s najvećim in-degree može biti centar attention u mreži.')
```
