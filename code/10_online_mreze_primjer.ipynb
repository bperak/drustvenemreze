Koristimo isključivo **simulirane** podatke (nema poziva na stvarni API, nema stvarnih korisnika).

## 1. Epistemologija: što graf *ne zna*

Prije brojeva kratko podsjetimo: **graf nije kultura**. On je model ograničen na ono što platforma bilježi i što vi uključite u istraživanje. Like, follow ili retweet u grafu **ne** govore sami po sebi o motivima, klasnoj poziciji, šali, ironiji ni o sadržaju koji je obrisan. Zato u radu uvijek navodimo **definiciju brida** i **vremenski okvir** — posebno kada rezultate čitamo iz perspektive **kulturalnih studija** (vidljivost, moć platforme, cirkulacija).

## 2. Usmjerena mreža: tko koga „prati”

Usmjereni graf (`DiGraph`) prikladan je kada brid ima smjer: **A prati B** nije isto što i **B prati A**. U interpretaciji to može značiti **pažnja** ili **hijerarhija** (npr. mnogi prate jednu instituciju; obrnuto ne mora vrijediti).

```python
import networkx as nx

# Mock: čvorovi su anonimni korisnici; usmjerena veza "A prati B"
D = nx.DiGraph()
D.add_edges_from(
    [
        ("u1", "u2"),
        ("u1", "u3"),
        ("u2", "u1"),
        ("u3", "u1"),
        ("u4", "u1"),
        ("u4", "u2"),
    ]
)

in_deg = dict(D.in_degree())
out_deg = dict(D.out_degree())
print("In-degree (broj sljedbenika / tko je 'praćen'):", in_deg)
print("Out-degree (broj praćenih računa):", out_deg)

# PageRank na usmjerenom grafu često se koristi kao proxy za "važnost" u mreži pažnje
pr = nx.pagerank(D, alpha=0.85)
print("PageRank (mock interpretacija — 'vidljivost' u ovoj mini-mreži):", pr)
```

**Kulturno čitanje (ilustrativno):** ako su `u1` i `u2` npr. dva javna diskursa koja se **uzajamno prate**, mreža sugerira **uzajamnu vidljivost**; ako `u4` samo prati druge bez da ga netko prati u uzorku, u ovom **rezanom** grafu on/ona je **periferan** — što u stvarnosti može značiti i tihu publiku, i bota, i kratkotrajni račun. Zato se kvalitativno provjerava **tko** stoji iza čvora.

## 3. Nesmjerena mreža oko kulturnog događaja (mock scenarij)

Pretpostavimo istraživačko pitanje: *tko se u javnom dijelu platforme pojavljuje u **istoj mreži interakcija** oko hipotetskog kulturnog festivala?*  
Bridovi u primjeru su **zajednički engagement** (npr. retweet ili mention u istom intervalu — ovdje već **pojednostavljeno** kao „povezani u istoj mock strukturi”).

Čvorovi su **tipovi aktera**, ne stvarne osobe.

```python
from networkx.algorithms.community import greedy_modularity_communities

G = nx.Graph()
# Lokalni krug oko festivala (gusto)
G.add_edges_from(
    [
        ("festival", "lokalni_portal"),
        ("festival", "volonter_A"),
        ("festival", "volonter_B"),
        ("lokalni_portal", "volonter_A"),
        ("lokalni_portal", "volonter_B"),
        ("volonter_A", "volonter_B"),
    ]
)
# Krug nacionalnih medija i institucije (gusto)
G.add_edges_from(
    [
        ("nacionalni_medij", "kriticar"),
        ("nacionalni_medij", "muzej_partner"),
        ("kriticar", "muzej_partner"),
    ]
)
# Jedan "most" između lokalnog i šireg polja javnosti
G.add_edge("festival", "nacionalni_medij")

deg = nx.degree_centrality(G)
print("Degree centrality (udio mogućih veza u ovoj mini-mreži):")
for node, v in sorted(deg.items(), key=lambda x: -x[1]):
    print(f"  {node}: {v:.3f}")

partitions = sorted(greedy_modularity_communities(G), key=len, reverse=True)
print("\nZajednice (greedy modularnost — ilustracija, ne 'istina' o terenu):")
for i, block in enumerate(partitions, 1):
    print(f"  Zajednica {i}: {sorted(block)}")
```

**Kulturno tumačenje:** algoritam je našao grupe **prema gustoći veza**, ne prema „zajedničkom identitetu”. Ipak, takav rezultat može poslužiti kao **hipoteza** za kvalitativno čitanje: postoji li u stvarnim podacima **odvojenost** između lokalnog organizacijskog kruga i kruga nacionalnog medijskog diskursa? **Most** (`festival` — `nacionalni_medij`) u kulturnim studijama možemo povezati s pitanjem **kako lokalni događaj ulazi u širu javnost** — ali samo uz analizu sadržaja i konteksta kampanje.

## 4. (Opcionalno) Vizualizacija

Raspored čvorova na crtežu **nije** inherentno svojstvo mreže — layout je estetski i didaktički izbor.

```python
import matplotlib.pyplot as plt

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8, 5))
nx.draw_networkx(
    G,
    pos,
    with_labels=True,
    node_color="#4a90d9",
    node_size=1200,
    font_size=9,
    font_color="white",
    edge_color="#555555",
    width=1.5,
)
plt.axis("off")
plt.title("Mock: nesmjerena mreža oko kulturnog događaja")
plt.tight_layout()
plt.show()
```
