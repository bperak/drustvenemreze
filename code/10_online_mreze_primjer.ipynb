Simulirana mreža "sljedbenika" (usmjerena) – bez poziva na stvarni API.

```python
import networkx as nx
# Mock: korisnici i usmjerene veze "A prati B"
D = nx.DiGraph()
D.add_edges_from([('u1','u2'),('u1','u3'),('u2','u1'),('u3','u1'),('u4','u1'),('u4','u2')])
print('In-degree (broj sljedbenika):', dict(D.in_degree()))
print('Out-degree (broj praćenja):', dict(D.out_degree()))
```
