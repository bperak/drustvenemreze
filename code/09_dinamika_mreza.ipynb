# Dinamika mreže – primjer

**Poglavlje:** [9. Dinamika mreža i promjene u vremenu](../content/09_dinamika_mreza_vrijeme.md)

Usporedba dva "snimka" mreže (simulirano).

```python
import networkx as nx
# "Snapshot" 1 i 2: dodavanje novih veza
G1 = nx.Graph(); G1.add_edges_from([(1,2),(2,3)])
G2 = nx.Graph(); G2.add_edges_from([(1,2),(2,3),(3,4),(4,1)])
print('t1: broj veza', G1.number_of_edges(), ', gustoća', round(nx.density(G1),4))
print('t2: broj veza', G2.number_of_edges(), ', gustoća', round(nx.density(G2),4))
```
