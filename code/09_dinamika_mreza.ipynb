**Poglavlje:** [9. Dinamika mreža i promjene u vremenu](../content/09_dinamika_mreza_vrijeme.md)

U kulturalnim studijima **snimak** mreže nije samo tehnički artefakt: on je *rez* u povijesti odnosa — što je platforma, arhiv ili istraživački korpus u tom trenutku još "vidio". Ispod uspoređujemo dva jednostavna grafa kao **t₁** i **t₂**: istu jezgru suradnje proširujemo novim akterom i vezama, pa pratimo gustoću, preklapanje bridova (Jaccard), degree centralnost i grubu modularnost zajednica (Louvain).


```python
import networkx as nx
from networkx.algorithms import community


def jaccard_bridova(G_prije, G_poslije):
    'Jaccardov indeks preklapanja skupova bridova (neusmjereni grafovi).'
    norm = lambda G: {frozenset((u, v)) for u, v in G.edges()}
    A, B = norm(G_prije), norm(G_poslije)
    if not A and not B:
        return 1.0
    return len(A & B) / len(A | B)


# Snimak t1: tijesna suradnja tri aktera (npr. jezgro scene prije širenja događaja)
G1 = nx.Graph()
G1.add_edges_from([("Ana", "Bruno"), ("Bruno", "Carla")])

# Snimak t2: novi akter i dodatne veze (npr. novi projekt / premještaj pažnje)
G2 = nx.Graph()
G2.add_edges_from([
    ("Ana", "Bruno"), ("Bruno", "Carla"), ("Carla", "David"), ("David", "Ana"),
])

print("--- Osnovne mjere ---")
for ozn, G in [("t1", G1), ("t2", G2)]:
    print(f"{ozn}: čvorovi={G.number_of_nodes()}, bridovi={G.number_of_edges()}, gustoća={nx.density(G):.4f}")

print("Preklapanje bridova (Jaccard, t1 vs t2):", round(jaccard_bridova(G1, G2), 4))

print()
print("--- Degree centralnost ---")
deg1, deg2 = nx.degree_centrality(G1), nx.degree_centrality(G2)
for n in sorted(G2.nodes()):
    d1 = deg1.get(n, 0.0)
    d2 = deg2.get(n, 0.0)
    print(f"  {n}: t1={d1:.3f}  t2={d2:.3f}")

comms1 = community.louvain_communities(G1, seed=42)
comms2 = community.louvain_communities(G2, seed=42)
Q1 = community.modularity(G1, comms1)
Q2 = community.modularity(G2, comms2)
print()
print("--- Louvain (gruba slika zajednica) ---")
print(f"t1: broj zajednica={len(comms1)}, modularnost Q={Q1:.3f}")
print(f"t2: broj zajednica={len(comms2)}, modularnost Q={Q2:.3f}")
```
