Primjer učitavanja odgovora iz CSV-a (npr. iz Google Formsa) i građenja egocentrične mreže. Za puni primjer s GForms i GSheet v. **Perak_Network_2.ipynb** u repozitoriju.

```python
import pandas as pd
import networkx as nx

# Primjer: tablica s dva stupca - ispitanik i osoba koju su naveli (name generator)
# Ako imate CSV iz Google Formsa, učitajte: df = pd.read_csv('odgovori.csv')
df = pd.DataFrame({
    'ispitanik': ['A', 'A', 'A', 'B', 'B', 'C'],
    'kontakt':   ['B', 'C', 'D', 'A', 'C', 'A']
})

G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row['ispitanik'], row['kontakt'])

print('Broj čvorova:', G.number_of_nodes())
print('Broj veza:', G.number_of_edges())
print('Čvorovi:', list(G.nodes()))
```
