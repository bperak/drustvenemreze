Identifikacija čvorova s visokim stupnjem (potencijalni "influenceri" u mreži preporuka).

```python
import networkx as nx
G = nx.Graph()
G.add_edges_from([('A','B'),('A','C'),('A','D'),('B','C'),('B','D'),('C','D'),('D','E')])
deg = nx.degree_centrality(G)
top = sorted(deg.items(), key=lambda x: -x[1])[:3]
print('Čvorovi s najvišim degree (potencijalni influenceri):', top)
```
