Mala **mreža preporuka / poznanstva** oko proizvoda ili brenda: gusti klaster (A–B–C–D) i periferijski čvor **E** koji je s „jezgom” povezan samo preko **D**. U marketinškoj interpretaciji **D** može biti i „popularan” (najviše direktnih veza) i **broker** prema novoj podmreži — ali to je **heuristika strukture**, ne dokaz da će E usvojiti poruku ili kupiti proizvod.

U kulturalnim studijima ovakav primjer koristimo da pokažemo da **različite mjere centralnosti** ne moraju dati isti rang — i da „influencer” kao pojam ovisi o tome **što smatra vezom** (preporuka, follow, wom, suradnja).

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

# Mreža: klaster A,B,C,D (međusobno gust povezan) + E samo preko D
G = nx.Graph()
G.add_edges_from(
    [("A", "B"), ("A", "C"), ("A", "D"), ("B", "C"), ("B", "D"), ("C", "D"), ("D", "E")]
)

deg = nx.degree_centrality(G)
bet = nx.betweenness_centrality(G)
eig = nx.eigenvector_centrality(G, max_iter=500)

print("Degree centrality:", {k: round(v, 3) for k, v in deg.items()})
print("Betweenness:", {k: round(v, 3) for k, v in bet.items()})
print("Eigenvector:", {k: round(v, 3) for k, v in eig.items()})
```

**Kulturno čitanje.** **Degree** naglašava tko ima mnogo **direktnih** kontakata (široka „mreža poznanstava” ili intenzivna interakcija u klasteru). **Betweenness** ističe **posrednika** bez kojega bi periferija bila odrezana — u kampanjama to može biti račun koji povezuje dvije subkulture. **Eigenvector** daje veću težinu čvorovima povezanim s drugim „centralnim” čvorovima: analogija je **legitimitet kroz povezanost** s već etabliranim akterima u sceni. Za potpunu kulturološku interpretaciju treba sadržaj (ton, žanr, vrijednosti) i kontekst platforme (pogl. 10).

```python
df = pd.DataFrame(
    {"degree": deg, "betweenness": bet, "eigenvector": eig}
).round(3)
df = df.sort_values("degree", ascending=False)
display(df)

top_deg = max(deg, key=deg.get)
top_bet = max(bet, key=bet.get)
print(f"Najviši degree: {top_deg}; najveći betweenness: {top_bet}")
```

```python
# Vizualizacija: boja ∝ betweenness (istaknut most prema E)
pos = nx.spring_layout(G, seed=42)
node_color = [bet[n] for n in G.nodes()]
nx.draw_networkx(
    G,
    pos,
    with_labels=True,
    node_color=node_color,
    cmap=plt.cm.Blues,
    edgecolors="black",
    font_size=11,
    node_size=900,
)
plt.title("WOM / preporuke: boja čvora ∝ betweenness")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# Doseg u 2 koraka (ego radijus 2) — grub proxy za „koliko ljudi je u bliskom prstenu”
for n in sorted(G.nodes()):
    ego = nx.ego_graph(G, n, radius=2)
    print(f"Čvor {n}: |ego| (r=2) = {ego.number_of_nodes()}")
```
