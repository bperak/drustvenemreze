# Usmjerene i težinske mreže

**Poglavlje:** [5. Mreže usmjerene i mreže s težinama](../content/05_mreze_usmjerene_tezine.md)

```python
import networkx as nx
D = nx.DiGraph()
D.add_edge('A','B', weight=2); D.add_edge('B','C', weight=1); D.add_edge('A','C', weight=3)
print('In-degree:', dict(D.in_degree()))
print('Out-degree:', dict(D.out_degree()))
print('Težinski in-degree:', dict(D.in_degree(weight='weight')))
```
