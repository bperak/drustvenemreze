# Socijalni kapital – primjer

**Poglavlje:** [4. Socijalni kapital i društvene mreže](../content/04_socijalni_kapital.md)

Ilustracija: gustoća i stupnjevi u maloj mreži kao proxy za „pristup” resursima.

```python
import networkx as nx
G = nx.karate_club_graph()
print('Gustoća:', nx.density(G))
print('Prosječni stupanj:', sum(dict(G.degree()).values())/G.number_of_nodes())
```

## Zaključak

Gustoća i prosječni stupanj daju osnovnu sliku „povezanosti” mreže; u istraživanjima socijalnog kapitala koriste se i ankete za validaciju.
