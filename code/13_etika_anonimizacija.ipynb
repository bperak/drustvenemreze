Primjer: zamjena identifikatora (pseudonimizacija) pri izradi mreže za objavu.

```python
# Primjer pseudonimizacije: mapiranje imena na ID
original = ['Ana', 'Bruno', 'Carla', 'David']
mapping = {name: f'P{i+1}' for i, name in enumerate(original)}
print('Mapiranje (za objavu koristiti samo ID):', mapping)
# U grafu zamijeniti čvorove: G = nx.relabel_nodes(G, mapping)
```
